```python
import cv2
```

```python
gorsel1=cv2.imread("image.jpeg")
```

```python
import matplotlib.pyplot as plt
image1=cv2.cvtColor(gorsel1,cv2.COLOR_BGR2RGB)
plt.imshow(image1)
```
